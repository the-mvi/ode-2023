–ß—Ç–æ –≤–æ–æ–±—â–µ –æ–∑–Ω–∞—á–∞–µ—Ç –∑–∞–¥–∞—á–∞, –∫–æ—Ç–æ—Ä—É—é –ø—Ä–µ–¥—Å—Ç–æ–∏—Ç —Ä–µ—à–∏—Ç—å, –≤ –≥–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–æ–º —Å–º—ã—Å–ª–µ? –û–∫–∞–∑—ã–≤–∞–µ—Ç—Å—è, —á—Ç–æ —Ç—É—Ç –ø—Ä–∏–º–µ–Ω—è–µ—Ç—Å—è —Ç–∞–∫ –Ω–∞–∑—ã–≤–∞–µ–º—ã–π "–º–µ—Ç–æ–¥ —Å—Ç—Ä–µ–ª—å–±", –∫–æ—Ç–æ—Ä—ã–π –∏ –ø–æ–∑–≤–æ–ª—è–µ—Ç –Ω–∞–π—Ç–∏ –Ω–µ–æ–±—Ö–æ–¥–∏–º—É—é h. h - —ç—Ç–æ –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è, –∫–æ—Ç–æ—Ä–∞—è –∑–∞–¥–∞–µ—Ç —É–≥–æ–ª –Ω–∞–∫–ª–æ–Ω–∞ "–ø—É—à–∫–∏", —Ç—Ä–∞–µ–∫—Ç–æ—Ä–∏—è —è–¥—Ä–∞ –∫–æ—Ç–æ—Ä–æ–π —è–≤–ª—è–µ—Ç—Å—è –≥—Ä–∞—Ñ–∏–∫–æ–º, –æ–ø–∏—Å–∞–Ω–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–µ–π y(x), –ø–æ–ª—É—á–∞–µ–º–æ–π –∏–∑ –¥–∏—Ñ—Ñ–µ—Ä–µ–Ω—Ü–∏–∞–ª—å–Ω–æ–≥–æ —É—Ä–∞–≤–Ω–µ–Ω–∏—è –∏–∑ —É—Å–ª–æ–≤–∏—è. –¢–æ—á–Ω–æ –Ω–∞–π—Ç–∏ –≥—Ä–∞—Ñ–∏–∫ –Ω–µ –ø–æ–ª—É—á–∏—Ç—Å—è, –Ω–æ –º–æ–∂–Ω–æ —Å –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–π —Ç–æ—á–Ω–æ—Å—Ç—å—é —É–∑–Ω–∞—Ç—å, –∫–∞–∫ –æ–Ω –±—É–¥–µ—Ç —Å–µ–±—è –≤–≤–µ—Å—Ç–∏ —Å –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–∞, –æ–ø–∏—Å–∞–Ω–Ω—ã—Ö –≤—ã—à–µ. –ù–∞ —Ä–∏—Å—É–Ω–∫–µ –Ω–∏–∂–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∞ —Ä–∞–±–æ—Ç–∞ –ø—Ä–æ–≥—Ä–∞–º–º—ã. –ò–∑–Ω–∞—á–∞–ª—å–Ω–æ –≤—ã–±–∏—Ä–∞–µ—Ç—Å—è —Å–ª—É—á–∞–π–Ω—ã–π —É–≥–æ–ª –Ω–∞–∫–ª–æ–Ω–∞ (–ø–µ—Ä–≤–∞—è –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏–∏ y(x)). –ö–∞–∫ –≤–∏–¥–Ω–æ –∏–∑ —Ä–∏—Å—É–Ω–∫–∞, —É–≥–æ–ª –Ω–∞–∫–ª–æ–Ω–∞ –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –æ—Å—Ç—Ä—ã–π, –∏–∑-–∑–∞ —á–µ–≥–æ –≥—Ä–∞—Ñ–∏–∫ –≤ —Ç–æ—á–∫–µ 2 —É—Ö–æ–¥–∏—Ç —Å–ª–∏—à–∫–æ–º –≤—ã—Å–æ–∫–æ. –ù–µ–æ–±—Ö–æ–¥–∏–º–æ —Å —à–∞–≥–æ–º 0,1 c–ø—É—Å–∫–∞—Ç—å—Å—è, –ø–æ–∫–∞ –Ω–µ –Ω–∞–π–¥–µ—Ç—Å—è –±–ª–∏–∂–∞–π—â–∞—è —Ç–æ—á–∫–∞, –∫–æ–≥–¥–∞ –≥—Ä–∞—Ñ–∏–∫ –ø—Ä–æ—Ö–æ–¥–∏—Ç –Ω–∞–¥ —Ç–æ—á–∫–æ–π y$_b$ = 3 (–∂–µ–ª—Ç–æ-–∑–µ–ª–µ–Ω–∞—è –ª–∏–Ω–∏—è), –∏ –±–ª–∏–∂–∞—â–∞—è —Ç–æ—á–∫–∞, –∫–æ–≥–¥–∞ –≥—Ä–∞—Ñ–∏–∫ –ø—Ä–æ—Ö–æ–¥–∏—Ç –ø–æ–¥ —Ç–æ—á–∫–æ–π y$_b$ = 3 (–±–∏—Ä—é–∑–æ–≤–∞—è –ª–∏–Ω–∏—è). 



–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –¥–∞–Ω–Ω–∞—è —Ç–æ—á–∫–∞ –ø–æ–ø–∞–¥–∞–µ—Ç –≤ "–≤–∏–ª–∫—É" –º–µ–∂–¥—É —ç—Ç–∏–º–∏ –¥–≤—É–º—è —Ç–æ—á–∫–∞–º–∏ –∏ –º–æ–∂–Ω–æ –≤–æ—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è –æ–±—ã—á–Ω—ã–º –º–µ—Ç–æ–¥–æ–º –ø–æ–ª–æ–≤–∏–Ω–Ω–æ–≥–æ –¥–µ–ª–µ–Ω–∏—è –¥–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è h (—É–≥–ª–∞), –¥–ª—è –∫–æ—Ç–æ—Ä–æ–≥–æ –≥—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ –ø—Ä–æ–π–¥–µ—Ç —á–µ—Ä–µ–∑ —Ç–æ—á–∫—É y$_b$ = 3.






## –í—ã–ø–æ–ª–Ω–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã

```python
import math
import random
from decimal import Decimal
import matplotlib
import matplotlib.pyplot as plt


def euler_method(x0, y0, dy0):

    result = []
    euler_step = Decimal("0.0001")

    current_x = Decimal(str(x0))
    current_y = Decimal(str(y0))
    current_dy = Decimal(str(dy0))
    current_ddy = f(current_x, current_y)

    s_dy = Decimal(current_dy + euler_step * current_ddy)
    next_y = Decimal(current_y + (current_dy + s_dy) * euler_step / 2)
    next_ddy = f(current_x + euler_step, next_y)

    next_dy = Decimal(current_dy + (current_ddy + next_ddy) * euler_step / 2)

    next_y = Decimal(current_y + (next_dy + current_dy) * euler_step / 2)

    # print("current_x", current_x, "current_y", current_y, "current_dy", current_dy, "current_ddy", current_ddy, "s_dy", s_dy, "next_y", next_y, "next_ddy", next_ddy)

    result.append({'x': current_x, 'y': current_y})

    while current_x < 2:
        current_x += euler_step
        current_y = next_y
        current_ddy = next_ddy
        current_dy = next_dy

        s_dy = current_dy + euler_step * current_ddy
        next_y = Decimal(current_y + (current_dy + s_dy) * euler_step / 2)
        next_ddy = f(current_x + euler_step, next_y)

        next_dy = Decimal(current_dy + (current_ddy + next_ddy) * euler_step / 2)

        next_y = current_y + (next_dy + current_dy) * euler_step / 2

        result.append({'x': current_x, 'y': current_y})

    return result



def f(x, y):


    value = Decimal(x * y + 2)

    return value.sqrt()


xa = 0
xb = 2

ya = 3
yb = 3

h = Decimal(random.random())
step = Decimal('0.1')

h_left = h
h_right = h

y_h_left = 0
y_h_right = 0

fig, ax = plt.subplots()

while True:
    result = euler_method(xa, ya, h_left)
    yh = result[-1]['y']
    print(yh)
    if yh > yb:
        h_left -= step
        y_h_left = yh

        x = [point['x'] for point in result]
        y = [point['y'] for point in result]
        ax.plot(x, y)
    else:
        h_right = h_left + step
        y_h_right = euler_method(xa, ya, h_right)[-1]['y']
        break

m = 0

best_h = 0

while m < 100:
    middle_h = Decimal((h_right + h_left) / 2)
    result = euler_method(xa, ya, middle_h)
    y_middle_h = result[-1]['y']
    print(y_middle_h)

    if y_middle_h > yb:
        h_right = middle_h
        y_h_right = y_middle_h
        best_h = h_right
    else:
        h_left = middle_h
        y_h_left = y_middle_h
        best_h = h_left

    x = [point['x'] for point in result]
    y = [point['y'] for point in result]
    ax.plot(x, y)

    if Decimal(abs(y_h_right - y_h_left)) < Decimal(0.0001):
        break
    m += 1

result = euler_method(xa, ya, best_h)

fig1, ax1 = plt.subplots()
x = [point['x'] for point in result]
y = [point['y'] for point in result]
ax1.plot(x, y)
print("best y'()", best_h)

plt.show()


```
